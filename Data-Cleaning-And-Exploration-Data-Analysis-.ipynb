```python
# show the columns with less than 50% missing values
missing_values_in_number=gt.isnull().sum()
missing_values_in_percentage =  (gt.isnull().sum() /gt.shape[0]) * 100    
missing_values_in_percentage = round(missing_values_in_percentage, 2)

missing_value=pd.concat([missing_values_in_number,missing_values_in_percentage ],axis=1, keys= ["missing_values_in_number",
                                                                                  "missing_values_in_percentage" ])

missing_value.query("missing_values_in_percentage >0 and missing_values_in_percentage <50",)
```

# Data Cleaning 

```python
# drop all columns with more than 50% missing values
x=missing_value.query("missing_values_in_percentage >=50",)
r=x.index.to_list()
gt.drop(columns=r, inplace=True)
```

```python
# drop all the columns with little weight 
gt.drop(columns=["latitude","longitude", "targsubtype1", "natlty1", "guncertain1", "nperps", "nperpcap", "claimed", 
             "weapsubtype1", "nkill", "nkillus", "nkillter", "nwound", "nwoundte", "nwoundus", "ishostkid"], inplace=True)
```

```python
# drop the missing value the subset of the following columns
gt.dropna(subset=["provstate", "city", "summary", "doubtterr", "multiple", "targsubtype1_txt", "corp1",
        "target1", "weapdetail", "scite1", "weapsubtype1_txt",  "natlty1_txt", "specificity"], inplace=True, axis=0)
```

```python
gt.shape
```

```python
clean_gt = gt.copy()
df=clean_gt.drop(columns=["INT_LOG", "INT_IDEO", "INT_MISC", "INT_ANY", 'crit1', 'crit2', 'crit3', 
                       'specificity', 'vicinity'])
```

```python
# 
total_crime_per_year=df.iyear.value_counts().to_frame("Crime Rate per Year").nlargest(10,
                                                                             "Crime Rate per Year").sort_index()


total_crime_success_per_year=df.query("success==1")["iyear"].value_counts().to_frame().nlargest(10, "iyear").sort_index()


s=pd.merge(total_crime_per_year, total_crime_success_per_year, left_index=True,     right_index=True,)
```

```python
fig, ax = plt.subplots(figsize=(12,8), facecolor="grey")

s.plot(kind="area", y=["Crime Rate per Year"] , ax=ax, cmap='Dark2')
ax1 = ax.twiny()
s.plot(kind="area", y=["iyear",] , ax=ax1, cmap='cubehelix')
```

```python
fig, ax = plt.subplots(figsize=(12,8), facecolor="grey")

s.plot(kind="bar", y=["Crime Rate per Year", "iyear"] , ax=ax, cmap='Dark2')
plt.title("hh", loc='center')
plt.ylabel("Total Count", color="blue", fontsize=30)
plt.xlabel("Year")
ax.tick_params(axis="x", labelrotation=0)
```
